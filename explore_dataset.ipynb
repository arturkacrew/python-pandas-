### Writing a function that automatically convert columns names to lower case and replace " " with "_"

```python

def clean_column_names(df):
    column_mapping = {col: col.lower().replace(' ', '_') for col in df.columns}
    df = df.rename(columns=column_mapping)
    return df
df_renamed = clean_column_names(df)
df_renamed
```

# Analysing the data

### What are the top countries that have the most number of successful bookings?

```python

top_countries_bookings_success = df_renamed \
    .query("is_canceled == 0") \
    .groupby('country',as_index=False) \
    .agg({'hotel':'count'}) \
    .sort_values('hotel', ascending=False)
```

```python
top_countries_bookings_success.head(10)
```

### What is the average nights bookings for hotels of different types for different week days?

```python

hotels_by_avg_nights_split = df_renamed \
    .groupby('hotel',as_index=False) \
    .agg({'stays_in_weekend_nights': 'mean','stays_in_week_nights':'mean'}) \
    .sort_values('hotel', ascending=False)
hotels_by_avg_nights_split
```

### What is the average length of stay (in nights) for each type of hotel, taking into account both weekend and weekday nights?


```python
hotels_by_avg_nights_unite = df_renamed \
    .groupby('hotel', as_index=False) \
    .agg({'stays_in_weekend_nights': 'mean', 'stays_in_week_nights': 'mean'}) \
    .assign(avg_nights=lambda x: x['stays_in_weekend_nights'] + x['stays_in_week_nights']) \
    .drop(['stays_in_weekend_nights', 'stays_in_week_nights'], axis=1) \
    .sort_values('hotel', ascending=False)
hotels_by_avg_nights_unite

```

### Analyzing dates of planned arrival. What are the top month with success bookings for 2016 by hotel? Do they changed for 2017? 


```python
booking_trends_by_month = df_renamed \
    .query("is_canceled == 0") \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .agg({'hotel': 'count'}) \
    .rename(columns={'hotel': 'successful_bookings'}) \
    .reset_index() \
    .sort_values('successful_bookings', ascending=False)
booking_trends_by_month
```

```python
#making pivot table to answer the questions above. 
pivot_table = booking_trends_by_month.pivot_table(values='successful_bookings', 
                                                  index='arrival_date_month', 
                                                  columns='arrival_date_year', 
                                                  aggfunc='sum')

# Calculate the sum of successful bookings for each month
sums_by_month = pivot_table.sum(axis=1)

# Sort the pivot table by the sums of successful bookings in descending order
pivot_table = pivot_table.loc[sums_by_month.sort_values(ascending=False).index]

pivot_table
```

### Is there any pattern between having kids and cancelation?


```python
#making table with total calculations
value_counts = df_renamed[['has_kids', 'is_canceled', 'arrival_date_year']].value_counts().reset_index(name='count')
value_counts
```

```python
#making pivot table to easily answer the question. 
pivot_table = pd.pivot_table(value_counts, values='count', index=['has_kids', 'arrival_date_year'], columns='is_canceled', fill_value=0)
pivot_table['percentage_canceled'] = ((pivot_table[1] / (pivot_table[0] + pivot_table[1])) * 100).round(2)
pivot_table
```
